```python
# Bu alanı çalıştırarak  verisetinin data isimli değişkene atanmasını sağlayınız.  
import pandas as pd 
import numpy as np

data = pd.read_csv('https://s3-eu-west-1.amazonaws.com/shanebucket/downloads/uk-500.csv')
data['numara'] = [np.random.randint(0,1000) for x in range(data.shape[0])]
 
data.head()
```

```python
# Soru 1: Sütun isimlerini sırasıyla isim, soyisim, firma_adi, adres, sehir, ilce, posta_kodu, tel1, tel2, eposta, web_sitesi, id olarak değiştiriniz. 
data = data.rename(columns={
    "first_name" : "isim", 
    "last_name" : "soyisim",
    "company_name" : "firma_adi", 
    "address" : "adres", 
    "city" : "sehir", 
    "county" : "ilce",
    "postal" : "posta_kodu", 
    "phone1" : "tel1",
    "phone2" : "tel2",
    "email" : "eposta",
    "web" : "web_sitesi",
    "numara" : "id"
})
```

```python
# Soru 2: numara isimli sütunu isim sütununun soluna taşıyınız. 
sutunlar = list(data.columns) #ipucu: sutunlar listesi uzerinde degisiklikler yaparak data = data[sutunlar] seklinde cozebilirsiniz. 
data = data[["id","isim", "soyisim", "firma_adi", "adres", "sehir", "ilce", "posta_kodu", "tel1", "tel2", "eposta", "web_sitesi"]]
data
```

```python
# Soru 3: Dataframe'in 1., 4., 7., 25. satır ve 1., 6.  7. sütunlarını getiriniz.
# (getirelecek dataframede 4 satır ve 3 sütun olacaktır)
data.iloc[[1,4,7,25],[1,6,7]] 
```

```python
# Soru 4: firma_adi sütunundaki son 20 kaydı getiriniz
data["firma_adi"].loc[:len(data)-20:-1]
```

```python
data.set_index("soyisim", inplace=True) # indexin düzenlenmesi (bu satırı yalnızca 1 kez çalıştırın)
```

```python
# Soru 5: indexleri düzenlenen dataframe üzerinde "Lin" indexli satırı getiriniz
data.loc["Lin",:]
```

```python
# Soru 6: Satırları'Andrade' ve 'Veness', sütunları ise sehir ve eposta arasindaki tüm sütunlar olan veri çerçevesini (dataframe) getiriniz.
data.loc[["Andrade", "Veness"], "sehir" : "eposta"]
```

```python
# Soru 7: İsmi Antonio olan kişilerin eposta adreslerini getiriniz. 
data.loc[data["isim"] == "Antonio","eposta"]
```
